
Based on the inspection, none of the dataframes have missing values.

The next step is to merge the relevant dataframes and handle categorical variables.DataFrames are merged to combine data from multiple sources into a single, unified dataset based on shared columns or indices.

The `orders_df` and `payment_df` will be merged on `order_id`, and then this can be merged with `customers_df` on `customer_id` to get customer information for each order. the combined dataframe with `customer_df` can now be merged with `shipment_df` can  The `transaction_status` in `payment_df` will be the target variable, and `payment_method` is a categorical feature that needs encoding( because logistic regression and random forest can not handle text or categorical strings directly) The `order_date`, `shipment_date` and `delivery_dat` in `orders_df` and `shipments_df` should be converted to datetime objects.








```python
# Merging orders and payment dataframes
orders_payment_df = pd.merge(orders_df, payment_df, on='order_id', how='left')

# Merging the combined dataframe with customers dataframe
orders_payment_customer_df = pd.merge(orders_payment_df, customers_df, on='customer_id', how='left')

# Merging the combined dataframe with shipments dataframe
orders_payment_customer_df = pd.merge(orders_payment_customer_df, shipments_df, on='order_id', how='left')


# Convert date columns to datetime objects
orders_payment_customer_df['order_date'] = pd.to_datetime(orders_payment_customer_df['order_date'])
orders_payment_customer_df['shipment_date'] = pd.to_datetime(orders_payment_customer_df['shipment_date'])
orders_payment_customer_df['delivery_date'] = pd.to_datetime(orders_payment_customer_df['delivery_date'])

# Ecoding payment_method
payment_method_mapping = {'Credit Card': 0, 'PayPal': 1, 'Bank Transfer': 2}
orders_payment_customer_df['payment_method_encoded'] = orders_payment_customer_df['payment_method'].map(payment_method_mapping)

# Displaying  the first few rows and data types of the resulting merged DataFrame
print("Merged DataFrame head:")
display(orders_payment_customer_df.head())
print("\nMerged DataFrame info:")
orders_payment_customer_df.info()
```

The data has been merged and date columns converted. The 'payment_method' column has been encoded and dropped. The 'transaction_status' column is the target variable and is currently an object type. This needs to be converted to a numerical representation for modeling. The remaining object type columns are identifiers, which are not suitable for direct use in the models and can be dropped.



```python
# custom mapping the target variable
status_mapping = {'Completed': 0, 'Failed': 1, 'Pending': 2}
orders_payment_customer_df['transaction_status_encoded_custom'] = orders_payment_customer_df['transaction_status'].map(status_mapping)

# dropping object type columns that are identifiers
columns_to_drop = ['first_name', 'last_name', 'address', 'email', 'phone_number', 'carrier', 'tracking_number', 'shipment_status', 'payment_method', 'transaction_status']
orders_payment_customer_df = orders_payment_customer_df.drop(columns=columns_to_drop)


# Displaying  the first few rows and data types of the new DataFrame
print("Final DataFrame head after custom encoding and dropping columns:")
display(orders_payment_customer_df.head())
print("\nFinal DataFrame info after custom encoding and dropping columns:")
orders_payment_customer_df.info()
```
